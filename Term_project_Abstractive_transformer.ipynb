```python
from transformers import pipeline
import os

```

```python
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
```

```python
summarizer_bart = pipeline("summarization")
```

```python
summarizer = pipeline("summarization", model="t5-base", tokenizer="t5-base", framework="tf")
```

```python
text = """A 28-year-old Bengaluru man, Vaishak Seraphim, builds clocks that turn anti-clockwise. A marketing professional by day, Vaishak dismantles clocks and alters the mechanism so that the hands of the clocks move in the anti-clockwise direction. He claims to have been inspired by a similar clock that his friend had seen at a cafe earlier."""
```

```python
summary_text = summarizer(text, max_length=20, min_length=5, do_sample=False)[0]['summary_text']
print(summary_text)
```

```python
summary_text_bart = summarizer_bart(text, max_length=20, min_length=5, do_sample=False)[0]['summary_text']
print(summary_text_bart)
```
